### variables

```python
val = genres['valence']
dance = genres['danceability']
ts = genres['time_signature']
dur = genres['duration_ms']
loud = genres['loudness']
live = genres['liveness']
energy = genres['energy']
gen = genres['genre']
acoust = genres['acousticness']
```

```python
matrix = np.corrcoef(val, ts)
print(matrix)
```

```python
sel_columns = [loud, dance, acoust, energy, live]
```

```python
correlation_matrix = genres.corr()
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sn.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Covariance Chart")
plt.show()
```

```python
abbrev_genres = pd.DataFrame(sel_columns)
```

```python
correlation_matrix = abbrev_genres.corr()
plt.figure(figsize=(5, 4))  # Adjust the figure size as needed
sn.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Covariance Chart")
plt.show()
```

```python
sn.boxplot(x='genre', y='energy', data=genres)
plt.xticks(rotation=90)
plt.title('Genre vs Energy')
plt.xlabel('Genre')
plt.ylabel('Energy')
plt.show()
```

```python
sn.boxplot(x='genre', y=loud, data=genres)
plt.xticks(rotation=90)
plt.title('Genre vs Loudness')
plt.xlabel('Genre')
plt.ylabel('Loudness')
plt.show()
```

### What Genre of music has a higher density of dancability songs?

```python
sn.boxplot(x='genre', y=dance, data=genres)
plt.xticks(rotation=90)
plt.title('Genre vs Danceability')
plt.xlabel('Genre')
plt.ylabel('Danceability')
plt.show()
```

### What is the average duration by genre?

```python
grouped = genres.groupby('genre').apply(lambda x: x.corr())

# Calculate mean correlation across genres
mean_corr_across_genres = grouped.groupby('genre').mean()

# Plot aggregated correlation matrix for all genres
plt.figure(figsize=(10, 8))
sn.heatmap(mean_corr_across_genres, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mean Correlation Matrix')
plt.show()
```
